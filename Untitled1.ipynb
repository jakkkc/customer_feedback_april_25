```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data with proper header handling
def load_data(filepath, month):
    # First find where the header row starts
    with open(filepath, 'r') as f:
        lines = f.readlines()
        for i, line in enumerate(lines):
            if 'Excellent' in line:  # Look for the header line
                header_row = i
                break
        else:
            header_row = 0  # Default if not found
    
    # Load with correct header
    df = pd.read_csv(filepath, skiprows=header_row)
    df['Month'] = month
    return df

# Reload all data
jan = load_data('sheets/jan25.csv', 'January')
feb = load_data('sheets/feb25.csv', 'February')
mar = load_data('sheets/march25.csv', 'March')
apr = load_data('sheets/apr25.csv', 'April')

# Combine and clean
all_data = pd.concat([jan, feb, mar, apr], ignore_index=True)
all_data = all_data.dropna(how='all')
all_data = all_data[all_data.iloc[:, 0] != '']

all_data.head()
```

```python
print("Columns in data:", all_data.columns.tolist())
```

```python
# Clean column names by stripping whitespace
all_data.columns = all_data.columns.str.strip()

# Rename first column if needed
all_data = all_data.rename(columns={all_data.columns[0]: 'Category'})

# Now try the rating columns again
rating_cols = ['Excellent', 'V.Good', 'Good', 'Average', 'Poor']
for col in rating_cols:
    all_data[col] = pd.to_numeric(all_data[col], errors='coerce').fillna(0)
```

```python
# Find all columns that contain rating data
rating_cols = [col for col in all_data.columns 
              if any(x in col for x in ['Excellent', 'V.Good', 'Good', 'Average', 'Poor'])]

# Map to standard names
col_mapping = {
    'Excellent': 'Excellent',
    'V.Good': 'V.Good', 
    'Good': 'Good',
    'Average': 'Average',
    'Poor': 'Poor'
}

# Rename columns
for col in rating_cols:
    for standard_name, variants in col_mapping.items():
        if standard_name in col:
            all_data = all_data.rename(columns={col: standard_name})
            break
```

```python
# Verify we have the expected columns
print("Columns after cleaning:", all_data.columns.tolist())
print("\nSample data:")
print(all_data.head())
print(all_data.columns.tolist())
```

```python
# List of rating columns (adjust if needed)
rating_cols = ['Excellent', 'V.Good', 'Good', 'Average', 'Poor']

# Convert ratings to numbers (fill missing with 0)
for col in rating_cols:
    all_data[col] = pd.to_numeric(all_data[col], errors='coerce').fillna(0)

# Calculate a weighted score (Excellent=5, Poor=1)
weights = [5, 4, 3, 2, 1]  # Weight for each rating
all_data['Score'] = (all_data[rating_cols] * weights).sum(axis=1) / all_data['Total']

print("\nData with calculated scores:")
print(all_data[['Category', 'Month', 'Score']])
```

```python
# Average score per month
monthly_avg = all_data.groupby('Month')['Score'].mean().sort_values()

print("\nAverage Monthly Scores:")
print(monthly_avg)

# Plot monthly performance
import matplotlib.pyplot as plt

monthly_avg.plot(kind='bar', color='skyblue')
plt.title("Hotel Performance (Jan-Apr 2025)")
plt.ylabel("Average Score (1-5)")
plt.ylim(3, 5)  # Set y-axis range for better visualization
plt.show()
```

```python
# Average score per category
category_avg = all_data.groupby('Category')['Score'].mean().sort_values(ascending=False)

print("\nTop 5 Categories:")
print(category_avg.head(5))

print("\nBottom 5 Categories:")
print(category_avg.tail(5))

# Plot best and worst categories
category_avg.head(5).plot(kind='barh', color='green', title="Top 5 Categories")
plt.show()

category_avg.tail(5).plot(kind='barh', color='red', title="Bottom 5 Categories")
plt.show()
```

```python
wifi_data = all_data[all_data['Category'] == 'Wi-Fi Connectivity']

print("\nWi-Fi Scores Over Months:")
print(wifi_data[['Month', 'Score']])

# Plot Wi-Fi trend
wifi_data.groupby('Month')['Score'].mean().plot(
    kind='line', marker='o', color='red', title="Wi-Fi Performance Trend"
)
plt.ylabel("Score (1-5)")
plt.ylim(1, 5)
plt.show()
```

```python
# Restaurant categories
restaurant_cats = ['Quality of Food', 'Variety of Food', 'Food Presentation', 'Quality of Service']
restaurant_data = all_data[all_data['Category'].isin(restaurant_cats)]

# Bar categories
bar_cats = ['Quality of beverages', 'Variety of beverages', 'Pricing', 'Quality of Service']
bar_data = all_data[all_data['Category'].isin(bar_cats)]

# Compare average scores
print("\nRestaurant Avg Score:", restaurant_data['Score'].mean())
print("Bar Avg Score:", bar_data['Score'].mean())

# Plot trends
restaurant_data.groupby('Month')['Score'].mean().plot(
    kind='line', marker='o', label='Restaurant'
)
bar_data.groupby('Month')['Score'].mean().plot(
    kind='line', marker='o', label='Bar'
)
plt.title("Restaurant vs Bar Performance")
plt.ylabel("Score (1-5)")
plt.legend()
plt.show()
```
